```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
df.info()
```

```python
profit_by_category = df.groupby("Category")["Profit"].sum()
profit_by_category
```

```python
profit_by_category.plot(kind="bar")
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```python
# Insight
# Technology category generates the highest profit, while Furniture shows lower profitability.
```

```python
profit_by_subcategory = (
    df.groupby("Sub-Category")["Profit"]
    .sum()
    .sort_values(ascending=False)
)

profit_by_subcategory.head(10)
```

```python
profit_by_subcategory.head(10).plot(kind="bar")
plt.title("Top 10 Sub-Categories by Profit")
plt.xlabel("Sub-Category")
plt.ylabel("Profit")
plt.show()
```

```python
profit_by_region = (
    df.groupby("Region")["Profit"]
    .sum()
    .sort_values(ascending=False)
)

profit_by_region
```

```python
profit_by_region.plot(kind="bar")
plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```

```python
#   Key Insights
# - A small number of sub-categories generate most of the profit
# - Profit distribution varies significantly across regions
# - Some categories show high sales but relatively low profit
```
